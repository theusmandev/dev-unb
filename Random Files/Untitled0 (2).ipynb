```python
# ================================================================
#   FACEBOOK ALBUM DOWNLOADER ‚Äî FULL WORKING CODE (GOOGLE COLAB)
#   Includes: Chrome installation, Chromedriver, Selenium, Scraper
#   Works in 2025 ‚Äî Works for public & friends-only albums
# ================================================================

# ---------------------------
# INSTALL CHROME + DRIVER
# ---------------------------

!apt-get update
!apt-get install -y wget unzip > /dev/null

# Install Chrome
!wget -O chrome.deb https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb
!apt-get install -y ./chrome.deb > /dev/null

# Find Chrome version
import subprocess
version = subprocess.check_output(["google-chrome", "--version"]).decode()
chrome_version = version.strip().split()[-1]
major = chrome_version.split('.')[0]

print("Installed Chrome Version:", chrome_version)

# Install matching ChromeDriver
driver_url = f"https://storage.googleapis.com/chrome-for-testing-public/{chrome_version}/linux64/chromedriver-linux64.zip"
!wget -O /tmp/chromedriver.zip {driver_url}
!unzip /tmp/chromedriver.zip -d /usr/bin/
!chmod +x /usr/bin/chromedriver-linux64/chromedriver

print("Chrome & ChromeDriver Installed Successfully ‚úî")

# ---------------------------
# IMPORT LIBRARIES
# ---------------------------

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import requests, shutil, time, os


# ---------------------------
# SETUP SELENIUM
# ---------------------------

def get_driver():
    options = Options()
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--no-sandbox")
    options.add_argument("--start-maximized")
    # DO NOT USE HEADLESS ‚Äî Facebook blocks login!
    # options.add_argument("--headless")

    return webdriver.Chrome(
        service=Service("/usr/bin/chromedriver-linux64/chromedriver"),
        options=options
    )


# ---------------------------
# HELPER FUNCTIONS
# ---------------------------

def clean_filename(name):
    return (name.replace("/", "_")
                .replace(":", "-")
                .replace("|", "")
                .replace("\n", " ")
                .strip())

def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)

def download_image(url, output_path):
    try:
        r = requests.get(url, stream=True)
        if r.status_code == 200:
            with open(output_path, "wb") as f:
                shutil.copyfileobj(r.raw, f)
        else:
            print("‚ùå Failed:", url)
    except Exception as e:
        print("‚ö† Error:", e)

def scroll_to_end(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1.4)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height


def extract_image_urls(driver):
    soup = BeautifulSoup(driver.page_source, "html.parser")
    urls = []

    for img in soup.find_all("img"):
        src = img.get("src", "")
        if "scontent" in src:
            urls.append(src)

    # remove duplicates
    return list(dict.fromkeys(urls))


# ---------------------------
# MAIN FUNCTION
# ---------------------------

def download_facebook_album(album_url):
    print("üöÄ Launching Chrome...")
    driver = get_driver()

    # Open Facebook first
    driver.get("https://www.facebook.com")
    print("\n‚ö†Ô∏è  Please LOGIN to Facebook manually in the browser window\n")
    time.sleep(25)  # wait for login

    print("üìÅ Opening album...")
    driver.get(album_url)
    time.sleep(3)

    print("üìú Scrolling to load all photos...")
    scroll_to_end(driver)

    print("üîç Extracting image URLs...")
    urls = extract_image_urls(driver)
    print(f"‚úî Found {len(urls)} images")

    album_title = clean_filename(driver.title)
    folder = f"FacebookAlbum_{album_title}"
    create_folder(folder)
    print("üìÇ Saving images to:", folder)

    for i, url in enumerate(urls, 1):
        print(f"‚¨á Downloading {i}/{len(urls)}")
        download_image(url, f"{folder}/{i}.jpg")

    driver.quit()
    print("\nüéâ DONE! Album downloaded successfully.")
    print("üìÅ Saved in folder:", folder)


# ---------------------------
# RUN THE DOWNLOADER
# ---------------------------

# üëâ REPLACE THIS WITH YOUR FACEBOOK ALBUM LINK
download_facebook_album("https://www.facebook.com/media/set/?set=a.204948602187212&type=3")
```
