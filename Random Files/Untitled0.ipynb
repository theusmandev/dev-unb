```python
# Step 1: Google Drive کو Colab میں ماؤنٹ کریں
from google.colab import drive
drive.mount('/content/drive')

# Step 2: لائبریری انسٹال کریں
!pip install pdf2image

# Step 3: لائبریری امپورٹ کریں
import os
from pdf2image import convert_from_path

# Step 4: اپنے Google Drive والے فولڈر کا path دیں
input_folder = "/content/drive/MyDrive/your_folder"   # <-- یہاں اپنا فولڈر دیں
output_folder = "/content/drive/MyDrive/pdf_images"

os.makedirs(output_folder, exist_ok=True)

# Step 5: سارے پی ڈی ایف کے پہلے پیج کو PNG میں سیو کریں
for file in os.listdir(input_folder):
    if file.lower().endswith(".pdf"):
        pdf_path = os.path.join(input_folder, file)
        images = convert_from_path(pdf_path, dpi=200, first_page=1, last_page=1)

        # پی ڈی ایف کے نام کے ساتھ ہی PNG سیو کریں
        pdf_name = os.path.splitext(file)[0]
        output_path = os.path.join(output_folder, f"{pdf_name}.png")
        images[0].save(output_path, "PNG")

        print(f"Saved: {output_path}")
```
