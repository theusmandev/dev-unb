```python
import requests
from bs4 import BeautifulSoup

def get_blogger_labels_custom(url):
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    r = requests.get(url, headers=headers, timeout=20)
    r.raise_for_status()

    soup = BeautifulSoup(r.text, "html.parser")
    labels = []

    # label-container ke andar label-head Label
    container = soup.find("div", class_="label-container")
    if not container:
        return []

    label_div = container.find("div", class_="label-head")
    if not label_div:
        return []

    # aksar labels <a> tags me hotay hain
    for a in label_div.find_all("a"):
        text = a.get_text(strip=True)
        if text:
            labels.append(text)

    return labels


# ====== USE HERE ======
url = "https://www.urdunovelbanks.com/2022/09/nimra-ahmed-complete-biography.html"

labels = get_blogger_labels_custom(url)

print("ðŸ“Œ Labels Found:")
for l in labels:
    print("-", l)
```
