```python
# ==========================================
# UrduNovelBanks - RSS Label URL Extractor
# USING BLOGGER RSS (MOST RELIABLE)
# ==========================================

import requests
import xml.etree.ElementTree as ET
from google.colab import files

# üîí HARD CODED LABEL
LABEL = "Nemrah Ahmed novels"

BASE_RSS = "https://www.urdunovelbanks.com/feeds/posts/default"

def fetch_urls_from_rss():
    urls = []
    start_index = 1
    max_results = 150

    while True:
        rss_url = (
            f"{BASE_RSS}"
            f"?alt=rss"
            f"&max-results={max_results}"
            f"&start-index={start_index}"
            f"&category={LABEL.replace(' ', '%20')}"
        )

        response = requests.get(rss_url)
        if response.status_code != 200:
            break

        root = ET.fromstring(response.content)

        items = root.findall(".//item")
        if not items:
            break

        for item in items:
            link = item.find("link")
            if link is not None and link.text:
                urls.append(link.text.strip())

        start_index += max_results

    return urls


print("‚è≥ RSS feed se URLs fetch ho rahi hain...\n")

urls = fetch_urls_from_rss()

# ===============================
# SAVE TO TXT
# ===============================
file_name = "writers_all_novels_list_urls.txt"

with open(file_name, "w", encoding="utf-8") as f:
    for url in urls:
        f.write(url + "\n")

print(f"‚úÖ Total URLs found: {len(urls)}")
print("üì• TXT file auto-download ho rahi hai...")

files.download(file_name)
```
