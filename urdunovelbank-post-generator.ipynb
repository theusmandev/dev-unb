```python
# Colab code with 2 Copy buttons
from IPython.display import display
import ipywidgets as widgets

def generate_download_button(drive_url):
    # File ID extract karna
    if "id=" in drive_url:
        file_id = drive_url.split("id=")[-1]
    elif "file/d/" in drive_url:
        file_id = drive_url.split("file/d/")[1].split("/")[0]
    else:
        raise ValueError("‚ö†Ô∏è Invalid Google Drive URL")

    # Direct link
    direct_link = f"https://drive.google.com/uc?export=download&id={file_id}"

    # HTML Code
    html_code = f"""
<div style="display: flex; justify-content: center;">
    <a class="custom-download-btn" href="{direct_link}" rel="noopener noreferrer" target="_blank">Download</a>
</div>
"""
    return direct_link, html_code


# ---- Run here ----
pdf_link = input("Apna Google Drive PDF link paste karein: ")

direct_link, html_code = generate_download_button(pdf_link)

# Output nicely
print("‚úÖ Direct Download Link:")
print(direct_link)
print("\n‚úÖ HTML Code:")
print(html_code)

# Copy button (Direct link)
copy_link_button = widgets.Button(description="üìã Copy Direct Link")

def on_copy_link(b):
    from google.colab import output
    output.eval_js(f'navigator.clipboard.writeText(`{direct_link}`)')
    print("üëâ Direct Link copied to clipboard!")

copy_link_button.on_click(on_copy_link)

# Copy button (HTML code)
copy_html_button = widgets.Button(description="üìã Copy HTML Code")

def on_copy_html(b):
    from google.colab import output
    output.eval_js(f'navigator.clipboard.writeText(`{html_code}`)')
    print("üëâ HTML Code copied to clipboard!")

copy_html_button.on_click(on_copy_html)

# Show buttons
display(copy_link_button, copy_html_button)
```

```python
import re
import requests
import io
from pdf2image import convert_from_bytes
from google.colab import files

def pdf_page_to_png(drive_url, page_number, output_name="output_page.png"):
    # Extract file ID from Google Drive URL
    if "id=" in drive_url:
        file_id = drive_url.split("id=")[-1]
    elif "file/d/" in drive_url:
        file_id = drive_url.split("file/d/")[1].split("/")[0]
    else:
        raise ValueError("Invalid Google Drive link")

    # Download URL
    download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

    # Download PDF in memory
    response = requests.get(download_url)
    if response.status_code != 200:
        raise Exception("Failed to fetch PDF from Google Drive")
    pdf_bytes = io.BytesIO(response.content)

    # Convert specific page to PNG
    images = convert_from_bytes(pdf_bytes.read(), first_page=page_number, last_page=page_number)
    images[0].save(output_name, "PNG")

    # Download in Colab
    files.download(output_name)
    print(f"‚úÖ Page {page_number} saved as {output_name}")

# Example usage
drive_link = "https://drive.google.com/file/d/1AkEDLunYsDyGc4GquIZQuZtxYxDTx_t8/view?usp=drive_link"
page_number = 1
pdf_page_to_png(drive_link, page_number, "zameen_ke_ansoo_page1.png")
```
