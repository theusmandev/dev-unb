```python
# Step 1: Install PyPDF2
!pip install PyPDF2

# Step 2: Import Libraries
import os
import requests
from PyPDF2 import PdfReader, PdfWriter
from google.colab import drive

# Step 3: Mount Google Drive
drive.mount('/content/drive')

def extract_multiple_ranges_from_drive(drive_url, ranges, output_folder, base_name="output"):
    """
    Google Drive link se PDF download karta hai,
    multiple page ranges extract karke har ek alag PDF banata hai.
    """
    # --- Drive file ID nikalna ---
    if "id=" in drive_url:
        file_id = drive_url.split("id=")[-1]
    elif "/d/" in drive_url:
        file_id = drive_url.split("/d/")[1].split("/")[0]
    else:
        raise ValueError("Google Drive URL sahi nahi hai.")

    download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

    # --- Temp file banayenge ---
    temp_file = "temp_input.pdf"
    response = requests.get(download_url, stream=True)
    with open(temp_file, "wb") as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)

    # --- PDF read ---
    reader = PdfReader(temp_file)
    total_pages = len(reader.pages)
    print(f"ğŸ“„ Input PDF me total {total_pages} pages hain")

    # --- Output folder ensure karein ---
    os.makedirs(output_folder, exist_ok=True)
    output_files = []

    # --- Har range ke liye alag PDF ---
    for idx, (start_page, end_page) in enumerate(ranges, start=1):
        writer = PdfWriter()

        # End page adjust kar do agar PDF se zyada ho
        end_page = min(end_page, total_pages)

        if start_page > total_pages:
            print(f"âš ï¸ Range {start_page}-{end_page} skip ki gayi (start page zyada hai).")
            continue

        for i in range(start_page-1, end_page):  # 1-based page numbers
            writer.add_page(reader.pages[i])

        output_pdf = os.path.join(output_folder, f"{base_name}_part{idx}.pdf")
        with open(output_pdf, "wb") as f:
            writer.write(f)

        output_files.append(output_pdf)
        print(f"âœ… Saved: {output_pdf}")

    # --- Temp file delete ---
    os.remove(temp_file)

    return output_files


# ---- Example Run ----
# Apna PDF ka Google Drive link
drive_url = "https://drive.google.com/file/d/1nJZdae2LrQVHyowsKdShT4x_nQHRLwjz/view?usp=sharing"

# Apne Google Drive ke novels folder ka path
drive_folder = "/content/drive/MyDrive/novels"

# Afaanajat folder banayenge usi path me
output_folder = os.path.join(drive_folder, "afaanajat")

# Multiple ranges (Example: 20â€“50, 60â€“80, 100â€“120)
ranges = [(20, 50), (60, 80), (100, 120),(190,200)]

output_files = extract_multiple_ranges_from_drive(drive_url, ranges, output_folder)

print("ğŸ‰ All files saved:")
for f in output_files:
    print(f)
```

```python
# Step 1: Install PyPDF2
!pip install PyPDF2

# Step 2: Import Libraries
import os
import requests
from PyPDF2 import PdfReader, PdfWriter
from google.colab import drive

# Step 3: Mount Google Drive
drive.mount('/content/drive')

def extract_ranges_with_titles(drive_url, ranges, output_folder):
    """
    Google Drive link se PDF download karta hai,
    har range ke liye custom title ke sath alag PDF banata hai.
    ranges = [(start_page, end_page, "title1"), (start_page, end_page, "title2"), ...]
    """
    # --- Drive file ID nikalna ---
    if "id=" in drive_url:
        file_id = drive_url.split("id=")[-1]
    elif "/d/" in drive_url:
        file_id = drive_url.split("/d/")[1].split("/")[0]
    else:
        raise ValueError("Google Drive URL sahi nahi hai.")

    download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

    # --- Temp file banayenge ---
    temp_file = "temp_input.pdf"
    response = requests.get(download_url, stream=True)
    with open(temp_file, "wb") as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)

    # --- PDF read ---
    reader = PdfReader(temp_file)
    total_pages = len(reader.pages)
    print(f"ğŸ“„ Input PDF me total {total_pages} pages hain")

    # --- Output folder ensure karein ---
    os.makedirs(output_folder, exist_ok=True)
    output_files = []

    # --- Har range ke liye alag PDF with title ---
    for start_page, end_page, title in ranges:
        writer = PdfWriter()

        # End page adjust kar do agar PDF se zyada ho
        end_page = min(end_page, total_pages)

        if start_page > total_pages:
            print(f"âš ï¸ Range {start_page}-{end_page} skip ki gayi (start page zyada hai).")
            continue

        for i in range(start_page-1, end_page):  # 1-based page numbers
            writer.add_page(reader.pages[i])

        safe_title = title.replace(" ", "_")  # spaces ko underscores me convert kar lein
        output_pdf = os.path.join(output_folder, f"{safe_title}.pdf")

        with open(output_pdf, "wb") as f:
            writer.write(f)

        output_files.append(output_pdf)
        print(f"âœ… Saved: {output_pdf}")

    # --- Temp file delete ---
    os.remove(temp_file)

    return output_files


# ---- Example Run ----
# Apna PDF ka Google Drive link
drive_url = "https://drive.google.com/file/d/1looLN8yFkTw5lrxu6J7TuTnlrefbNscT/view?usp=sharing"

# Apne Google Drive ke novels folder ka path
drive_folder = "/content/drive/MyDrive/novels"

# Afaanajat folder banayenge usi path me
output_folder = os.path.join(drive_folder, "afaanajat")

# Multiple ranges with custom titles
ranges = [
    (27, 29, "Niki jai han by Rashida Riffat_a"),
    (30, 35, "Bara bay adab hun by Salma yasmeen_a"),
    (35, 46, "Parakh by Rukhsana Nigar Adnan_n"),

]

output_files = extract_ranges_with_titles(drive_url, ranges, output_folder)

print("ğŸ‰ All files saved:")
for f in output_files:
    print(f)
```
