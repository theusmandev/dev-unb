```python
import os
from google.colab import drive

# ðŸ”¹ Drive mount
drive.mount('/content/drive')

# ðŸ”¹ Google Drive folder path
target_folder = "/content/drive/MyDrive/urdunovelbank3/urdunovelbank3/Zainab Khan Novels"

# ðŸ”¹ Output TXT same folder mein
output_file = os.path.join(target_folder, "file_names.txt")

file_names = []

remove_text = "(www.urdunovelbanks.com)"

for file in os.listdir(target_folder):
    file_path = os.path.join(target_folder, file)

    if os.path.isfile(file_path):
        # ðŸ”¹ extension remove
        name_without_ext = os.path.splitext(file)[0]

        # ðŸ”¹ end se specific text remove
        if name_without_ext.endswith(remove_text):
            name_without_ext = name_without_ext.replace(remove_text, "").strip()

        file_names.append(name_without_ext)

# ðŸ”¹ TXT write
with open(output_file, "w", encoding="utf-8") as f:
    for name in file_names:
        f.write(name + "\n")

print("âœ… file_names.txt create ho gayi (extra text remove kar diya gaya)")
```
